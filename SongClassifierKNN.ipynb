```python
import pandas as pd
import numpy as np 
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

#Removing entries for genres with less volume and dropping insignificant columns
data = pd.read_csv('songs.csv')
data['title'].drop_duplicates()
data['top genre'].value_counts()
t = data['top genre'].value_counts()
filtered = data.groupby('top genre').filter(lambda x : len(x)>12)
filtered = filtered.drop(columns = ['target', 'artist', 'added'])
filtered = filtered.drop_duplicates(['title'])
filtered.shape
```

```python
#Map categorical variables to numeric

filtered['top genre'].value_counts()
filtered['top genre'] = filtered['top genre'].map({'desi pop': 0, 'dance pop': 1, 'hip hop': 2, 'emo rap': 3, 'conscious hip hop': 4})
filtered
```

```python
#train test split and scale the variables
scale = StandardScaler()
pd.set_option('display.max_rows', None)
x = filtered[list(filtered.columns[0:14])]
y = filtered[list(filtered.columns[1:2])]
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2, random_state = 1)
print(x_train)
x_train = x_train.drop(columns=['top genre'])
x_test = x_test.drop(columns=['top genre'])
print(x_train.head(98))
dataPoints = {}
x_train = x_train.reset_index()
x_test = x_test.reset_index()
for i in range(0,len(x_train)):
    title = x_train['title'][i]
    info = scale.fit_transform(x_train[list(x_train.columns[2:12])])
    dataPoints[i] = (title, np.array(info[i]))
dataPoints[0]
```

```python
from scipy import spatial

def ComputeDistance(a, b):
    dataA = a[1]
    dataB = b[1]
    similarityDistance = spatial.distance.cosine(dataA, dataB)
    return similarityDistance
    
ComputeDistance(dataPoints[11], dataPoints[2])

dataPoints[27]
```

```python
import operator

def getNeighbors(songID, K):
    distances = []
    for song in dataPoints:
        if (song != songID):
            dist = ComputeDistance(dataPoints[songID], dataPoints[song])
            distances.append((song, dist))
    distances.sort(key=operator.itemgetter(1))
    neighbors = []
    for x in range(K):
        neighbors.append(distances[x][0])
    return neighbors

K = 10
neighbors = getNeighbors(55, K)
for neighbor in neighbors:
    print (dataPoints[neighbor][0])
```
